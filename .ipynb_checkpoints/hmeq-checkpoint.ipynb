## Dataset summary
The data set HMEQ reports characteristics and delinquency information for 5,960 home equity loans. A home equity loan is a loan where the obligor uses the equity of his or her home as the underlying collateral. The data set has the following characteristics:

◾ BAD: 1 = applicant defaulted on loan or seriously delinquent; 0 = applicant paid loan

◾ LOAN: Amount of the loan request

◾ MORTDUE: Amount due on existing mortgage

◾ VALUE: Value of current property

◾ REASON: DebtCon = debt consolidation; HomeImp = home improvement

◾ JOB: Occupational categories

◾ YOJ: Years at present job

◾ DEROG: Number of major derogatory reports

◾ DELINQ: Number of delinquent credit lines

◾ CLAGE: Age of oldest credit line in months

◾ NINQ: Number of recent credit inquiries

◾ CLNO: Number of credit lines

◾ DEBTINC: Debt-to-income ratio

source: http://www.creditriskanalytics.net/datasets-private2.html

### Missing values

```python
df.isna().sum()
```

Due to the missing data present in most variables, multiple imputation is used, as in this case simple imputation methods (mean or median) could lead to biased estimators.

```python
mice_data = MICEData(df.select_dtypes(exclude=['object','category']))
mice_data.update_all()

df_imputed = mice_data.data
```

```python
df_combined = pd.concat([df_imputed, df.select_dtypes(include=['object','category'])], axis=1)
```

The remaining missing values in the categorical variables will be replaced with the most frequent category.

```python
df_combined['REASON'].fillna(df_combined['REASON'].mode()[0], inplace=True)
df_combined['JOB'].fillna(df_combined['JOB'].mode()[0], inplace=True)
```

## Customer segmentation

```python
X = df_combined.drop(['BAD'],axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = [] #Within-Cluster Sum of Squares
silhouette_scores = []
max_clusters = 10 

for n_clusters in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
    if n_clusters > 1:
        silhouette_scores.append(silhouette_score(X, kmeans.labels_))

plt.figure(figsize=(12, 6))
plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, max_clusters + 1))
plt.grid()
plt.show()
```

```python
if silhouette_scores:
    plt.figure(figsize=(12, 6))
    plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o', linestyle='--', color='orange')
    plt.title('Silhouette Score vs Number of Clustters')
    plt.xlabel('Number of clusters')
    plt.ylabel('Value of Silhouette Score')
    plt.xticks(range(2, max_clusters + 1))
    plt.grid()
    plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
optimal_clusters = 2

kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
kmeans.fit(X)
cluster_labels = kmeans.labels_

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(10, 8))
for cluster in range(optimal_clusters):
    plt.scatter(
        X_pca[cluster_labels == cluster, 0], 
        X_pca[cluster_labels == cluster, 1], 
        label=f'Cluster {cluster+1}', 
        alpha=0.6
    )

plt.title('Customer segmentation')
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
X['CLUSTER'] = cluster_labels
```

```python
df_segments = pd.concat([X,df_combined['BAD']],axis=1)
```

```python
df_segments[num_var] = scaler.inverse_transform(df_segments[num_var])
```

```python
cluster_target_crosstab = pd.crosstab(df_segments['CLUSTER'], df_segments['BAD'], normalize='index')
print(cluster_target_crosstab)
```

```python
cluster_target_crosstab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Good vs bad clients')
plt.ylabel('Udział')
plt.xlabel('Klaster')
plt.legend(title='BAD')
plt.show()
```

```python
cluster_summary = df_segments.groupby('CLUSTER')[num_var].mean()
target_summary = df_segments.groupby('CLUSTER')['BAD'].value_counts(normalize=True).unstack()

print("Summary between clusters:")
print(cluster_summary)
```

```python
target_summary
```

There are clear differences in the proportion of good and bad clients across the clusters. Cluster 1 contains a higher share of good clients. The characteristics of the clients in both clusters will be further analyzed.

```python
for var in continuous_num_var:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_segments, x='CLUSTER', y=var, palette='tab10')
    plt.title(f'Distribution for {var} between clusters')
    plt.xlabel('Cluster')
    plt.ylabel(var)
    plt.grid(axis='y', alpha=0.7)
    plt.show()
```

## Split and oversampling

```python
## Split
X = df_combined.drop(['BAD'],axis=1)
y = df_combined['BAD']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=df_combined['BAD'], random_state=42)
## Oversampling
y_train.value_counts(normalize=True)
smote = SMOTE(random_state=42)
 
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
y_train_smote.value_counts(normalize=True)
X_train_smote.shape
```

# Classification models

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform, loguniform
from sklearn.metrics import classification_report,confusion_matrix, ConfusionMatrixDisplay,roc_curve, auc, recall_score, roc_auc_score
```

```python
rf = RandomForestClassifier(
    n_jobs=-1,
    random_state=42,
)

rf_param_dist = {
    "n_estimators": randint(200, 1001),
    "max_depth": randint(3, 31),
    "min_samples_split": randint(2, 21),
    "min_samples_leaf": randint(1, 21),
    "max_features": np.linspace(1, X_train.shape[1],16).astype(int),
    "bootstrap": [True, False]
}
```

```python
rf_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=rf_param_dist,
    n_iter=50,
    scoring="roc_auc",     #PD
    cv=3,
    n_jobs=-1,
    random_state=42
)

rf_search.fit(X_train, y_train)
best_rf = rf_search.best_estimator_
print(rf_search.best_params_, rf_search.best_score_)
```

```python
y_pred_test_ = best_rf.predict(X_test)
recall_test = recall_score(y_test, y_pred_test_)
recall_test
```

```python
print(classification_report(y_test, y_pred_test_))
```

```python
y_pred_t_proba = best_rf.predict_proba(X_test)

fpr, tpr, thresholds = roc_curve(y_test, y_pred_t_proba[:,1])
rf_roc_auc = roc_auc_score(y_test, y_pred_t_proba[:,1])
roc_auc = auc(fpr, tpr)  

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label="Random model")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
cm = confusion_matrix(y_test,y_pred_test_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_rf.classes_)
disp.plot(cmap='Blues')  
plt.title("Confusion matrix")
plt.show()
```

```python
importances = best_rf.feature_importances_
std = np.std([tree.feature_importances_ for tree in best_rf.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

num_feat = 6
plt.figure(figsize=[15, 8])
plt.title("Feature importance")
plt.bar(
    range(num_feat)[:num_feat],
    importances[indices][:num_feat],
    color="skyblue",
    yerr=std[indices][:num_feat],
    align="center",
)
plt.xticks(range(num_feat)[:num_feat], X_train.columns[indices[:num_feat]])
plt.xlim([-1, num_feat])
plt.ylabel("Impurity reduction")
```

## Summary

```python
standard = [recall_test, gbm_recall_test]          
smote    = [smote_recall_test, smote_gbm_recall_test]

auc_standard = [rf_roc_auc, gbm_roc_auc]
auc_smote    = [smote_rf_roc_auc, smote_gbm_auc_roc]

labels = ["RF", "GBM"]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(9,6))
bars1 = plt.bar(x - width/2, standard, width, label="Standard Recall", color="tab:blue")
bars2 = plt.bar(x + width/2, smote, width, label="SMOTE Recall", color="tab:orange")

plt.xticks(x, labels)
plt.ylabel("Recall")
plt.title("Recall Comparison With ROC AUC Values")
plt.ylim(0, 1)

plt.grid(axis='y', linestyle='--', alpha=0.6)

for i, v in enumerate(standard):
    plt.text(x[i] - width/2, v + 0.02, f"{v:.3f}", ha='center', fontsize=10)
    plt.text(x[i] - width/2, 0.02, f"AUC={auc_standard[i]:.3f}", ha='center', color='white')

for i, v in enumerate(smote):
    plt.text(x[i] + width/2, v + 0.02, f"{v:.3f}", ha='center', fontsize=10)
    plt.text(x[i] + width/2, 0.02, f"AUC={auc_smote[i]:.3f}", ha='center', color='white')

plt.legend()
plt.show()
```

Overall, applying SMOTE consistently improves the models ability to correctly identify the positive class, with the strongest improvement observed in GBM. Additionally, the discriminatory power was also absolutely higher for the models trained after balancing the sample.
